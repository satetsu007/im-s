```python
import os
import pandas as pd
```

```python
os.chdir('/media/satetsu-gpu/Data/program/DL/anime/imas')
folder = "Dataset/02"
filename = os.listdir(folder)
idolname = [idol for idol in filename if idol[-4:] != ".jpg"]
idolname
```

```python
idol_list = []
file_list = []
for i, idol in enumerate(idolname):
    files = os.listdir(folder + "/" + idol)
    for file in files:
        idol_list.append([file, i])
        file_list.append(file)
```

```python
idol_dict = dict(idol_list)
```

```python
tmp = [0 for i in idolname]
```

```python
for file in file_list:
    tmp = [0 for i in idolname]
    flag = idol_dict[file]
    tmp[flag] = 1
    idol_dict[file] = tmp
```

```python
idol_df = pd.DataFrame.from_dict(idol_dict).T
```

```python
idol_df.columns = idolname
```

```python
idol_df.to_csv(folder + ".csv")
```
